```python
# Ãœyelik tÃ¼rÃ¼ne gÃ¶re ortalama zaman harcama miktarÄ±nÄ± hesaplama
kategori_zaman = data.groupby("abonoment_type")[["avg_time_in_gym"]].mean().reset_index()

# Grafik oluÅŸturma
plt.figure(figsize=(25, 6))
plt.bar(kategori_zaman["abonoment_type"], kategori_zaman["avg_time_in_gym"])
plt.xlabel("Ãœyelik TÃ¼rÃ¼")
plt.ylabel("Ortalama Salonda GeÃ§irilen SÃ¼re (dakika)")
plt.title("Ãœyelik TÃ¼rÃ¼ne GÃ¶re Ortalama Salonda GeÃ§irilen SÃ¼re")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='abonoment_type', y='avg_time_in_gym', data=data, palette="Set2")
plt.xlabel("Ãœyelik TÃ¼rÃ¼")
plt.ylabel("Salonda GeÃ§irilen SÃ¼re (dakika)")
plt.title("Ãœyelik TÃ¼rÃ¼ne GÃ¶re Salonda GeÃ§irilen SÃ¼relerin DaÄŸÄ±lÄ±mÄ±")
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='visit_per_week', y='avg_time_in_gym', hue='abonoment_type', palette='Set1', s=100, alpha=0.7)

plt.title("HaftalÄ±k Ziyaret SayÄ±sÄ± ile Salonda GeÃ§irilen SÃ¼re ArasÄ±ndaki Ä°liÅŸki")
plt.xlabel("HaftalÄ±k Ziyaret SayÄ±sÄ±")
plt.ylabel("Salonda GeÃ§irilen SÃ¼re (dakika)")
plt.legend(title='Ãœyelik TÃ¼rÃ¼')
plt.grid(True)
plt.show()
```

```python
data = data.dropna() #silme işlemi
```

```python
numeric_columns = data.select_dtypes(include=['float64', 'int64'])

data[numeric_columns.columns] = numeric_columns.fillna(numeric_columns.mean())
```

```python
# Üyelik türüne göre ortalama zaman harcama miktarını hesaplama
kategori_zaman = data.groupby("abonoment_type")[["avg_time_in_gym"]].mean().reset_index()

# Grafik oluşturma
plt.figure(figsize=(25, 6))
plt.bar(kategori_zaman["abonoment_type"], kategori_zaman["avg_time_in_gym"])
plt.xlabel("Üyelik Türü")
plt.ylabel("Ortalama Salonda Geçirilen Süre (dakika)")
plt.title("Üyelik Türüne Göre Ortalama Salonda Geçirilen Süre")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='abonoment_type', y='avg_time_in_gym', data=data, palette="Set2")
plt.xlabel("Üyelik Türü")
plt.ylabel("Salonda Geçirilen Süre (dakika)")
plt.title("Üyelik Türüne Göre Salonda Geçirilen Sürelerin Dağılımı")
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='visit_per_week', y='avg_time_in_gym', hue='abonoment_type', palette='Set1', s=100, alpha=0.7)

plt.title("Haftalık Ziyaret Sayısı ile Salonda Geçirilen Süre Arasındaki İlişki")
plt.xlabel("Haftalık Ziyaret Sayısı")
plt.ylabel("Salonda Geçirilen Süre (dakika)")
plt.legend(title='Üyelik Türü')
plt.grid(True)
plt.show()
```
